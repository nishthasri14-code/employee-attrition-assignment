## Outlier Detection using IQR Method

The Interquartile Range (IQR) method is used to detect outliers in the MonthlyIncome column.

Lower Bound = Q1 − 1.5 × IQR  
Upper Bound = Q3 + 1.5 × IQR

```python
# Handling Outliers using IQR Method
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['MonthlyIncome'] < lower) | (df['MonthlyIncome'] > upper)]

print("Number of outliers:", outliers.shape[0])
```

```python
# Removing outliers
df = df[(df['MonthlyIncome'] >= lower) & (df['MonthlyIncome'] <= upper)]
```

## Data Transformation

Data transformation is performed to prepare the dataset for machine learning algorithms.

### 1. Standardization

Standardization is applied to the **MonthlyIncome** feature using the StandardScaler.  
Different numerical features in a dataset can have different ranges. For example, income values may be very large compared to other numerical variables. If the data is not scaled, features with larger values can dominate the model.

Standardization converts the data so that it has a **mean of 0 and a standard deviation of 1**, which helps improve the performance and stability of machine learning algorithms.

### 2. Label Encoding

The **Attrition** column is a categorical variable containing values such as *Yes* and *No*. Machine learning algorithms require numerical input, so categorical values must be converted into numbers.

Label Encoding converts the categories into numerical values, for example:

No → 0  
Yes → 1  

This allows the model to process the categorical information during training.

```python
# Scaling (Standardization)
df = df.copy()

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df['MonthlyIncome_scaled'] = scaler.fit_transform(df[['MonthlyIncome']])
```

```python
# Encoding Categorical Variable
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['Attrition_encoded'] = encoder.fit_transform(df['Attrition'])
```
